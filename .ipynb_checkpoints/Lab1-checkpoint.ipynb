```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img=cv2.imread("descarga.jpg")
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## Problema 1

```python
def transformImg(img, color):
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzo = np.zeros((alto, ancho, 3))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            if(color==1):
                lienzo[i, j] = [blue, 0, 0]
            elif(color==2):
                lienzo[i, j] = [0, green, 0]
            elif(color==3):
                lienzo[i, j] = [0, 0, red]    
            elif(color==10):
                lienzo[i, j] = [0, green, red]      
            elif(color==20):
                lienzo[i, j] = [blue, green, 0]    
            elif(color==30):
                lienzo[i, j] = [blue, 0, red] 
    cv2.imwrite("monocromo.jpg", lienzo)
    monocromo = cv2.imread("monocromo.jpg")
    monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    plt.imshow(monocromo)
    plt.show()      
                
```

```python
transformImg(img,30)
```

## Problema 3

```python
def transformImg3(img):    
    plt.imshow(img)
    plt.show()
    alto = img.shape[0]
    ancho = img.shape[1]
    media_ponderadaR = np.zeros((alto, ancho, 1))
    media_ponderadaG = np.zeros((alto, ancho, 1))
    media_ponderadaB = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            media_ponderadaR[i, j] = int(0.114*red)
            media_ponderadaG[i, j] = int(0.587*green)
            media_ponderadaB[i, j] = int(0.299*blue)
    cv2.imwrite("gris_ponderadoR.jpg", media_ponderadaR)
    grisR = cv2.imread("gris_ponderadoR.jpg")
    grisR = cv2.cvtColor(grisR, cv2.COLOR_BGR2RGB)
    plt.imshow(grisR)
    plt.show()
    cv2.imwrite("gris_ponderadoG.jpg", media_ponderadaG)
    grisG = cv2.imread("gris_ponderadoG.jpg")
    grisG = cv2.cvtColor(grisG, cv2.COLOR_BGR2RGB)
    plt.imshow(grisG)
    plt.show()
    cv2.imwrite("gris_ponderadoB.jpg", media_ponderadaB)
    grisB = cv2.imread("gris_ponderadoB.jpg")
    grisB = cv2.cvtColor(grisB, cv2.COLOR_BGR2RGB)
    plt.imshow(grisB)
    plt.show()
```

```python
transformImg3(img)
```

## Problema 4

```python
def plotHistImg(img):
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzoR = np.zeros((alto, ancho, 3))
    lienzoG = np.zeros((alto, ancho, 3))
    lienzoB = np.zeros((alto, ancho, 3))
    lienzoGrey = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            lienzoR[i, j] = [0, 0, red]
            lienzoG[i, j] = [0, green, 0]
            lienzoB[i, j] = [blue, 0, 0]
            lienzoGrey[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
    cv2.imwrite("lienzoR.jpg", lienzoR)
    lienzoR = cv2.imread("lienzoR.jpg")
    lienzoR = cv2.cvtColor(lienzoR, cv2.COLOR_BGR2RGB)
    plt.hist(lienzoR.ravel(),256,[0,256])
    plt.show()
    
    cv2.imwrite("lienzoG.jpg", lienzoG)
    lienzoG = cv2.imread("lienzoG.jpg")
    lienzoG = cv2.cvtColor(lienzoG, cv2.COLOR_BGR2RGB)
    plt.hist(lienzoG.ravel(),256,[0,256])
    plt.show()
    
    cv2.imwrite("lienzoB.jpg", lienzoB)
    lienzoB = cv2.imread("lienzoB.jpg")
    lienzoB = cv2.cvtColor(lienzoB, cv2.COLOR_BGR2RGB)
    plt.hist(lienzoB.ravel(),256,[0,256])
    plt.show()
    
    cv2.imwrite("lienzoGrey.jpg", lienzoGrey)
    lienzoGrey = cv2.imread("lienzoGrey.jpg")
    lienzoGrey = cv2.cvtColor(lienzoGrey, cv2.COLOR_BGR2RGB)
    plt.hist(lienzoGrey.ravel(),256,[0,256])
    plt.show()
```

```python
plotHistImg(img)
```
